# 4.Soru

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
def f(x):
    return x**3-12*x

def g(x):
    return x**2
```

```python
x = sy.Symbol('x')
sy.solve(f(x) - g(x))
```

```python
x = np.linspace(-3,4,100)
plt.plot(x,f(x))
plt.plot(x,g(x))
plt.fill_between(x, f(x), color = "grey", alpha = 0.2, hatch = '|')
plt.fill_between(x, g(x), color = "yellow", alpha = 0.3, hatch = '-')
```
