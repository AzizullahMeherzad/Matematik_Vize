# 3.Soru

```python
import matplotlib.pyplot as plt
import numpy as np
import sympy as sy
```

```python
def f(x):
    return x**2/3

def g(x):
    return 4-(2/3)*x**2
```

```python
x = sy.Symbol('x')
sy.solve(f(x) - g(x))
```

```python

x = np.linspace(-2,2,100)
plt.plot(x,f(x))
plt.plot(x,g(x))
plt.fill_between(x, f(x), color = "grey", alpha = 0.2, hatch = '|')
plt.fill_between(x, g(x), color = "yellow", alpha = 0.3, hatch = '-')
```
